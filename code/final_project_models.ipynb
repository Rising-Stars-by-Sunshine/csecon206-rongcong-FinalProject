
# **1.Competitive Game**

**1.1 Game Environment:**
The Competitive game environment simulates real-life social networking platforms. In this virtual world, there are have players—Player A and Player B, representing two social media influencers (SMIs) with similar fan bases and content types. There is a clear competition between these two players as they strive to gain the favor of the audience and stand out on this virtual social media platform.

**1.2 Model 1 description:**

When SMIA and SMIB release their posts in no obvious order.

If both SMIA and SMIB choose not to release explicit advertisements, their payoff will be 5 each.
If both SMIA and SMIB choose to release explicit advertisements, their payoff will be 3 each.
If SMIA chooses to release explicit advertisements but SMIB chooses not to, SMIA's payoff will be 0, while SMIB's payoff will be 10.
If SMIB chooses to release explicit advertisements but SMIA chooses not to, SMIB's payoff will be 0, while SMIA's payoff will be 10.
And vice versa.




|     | no obvious ad | obvious ad
|  ----  | :----:  | :----: |
| no obvious ad  | 5, 5 | 10，0 |
| obvious ad  | 0，10 |  3，3 |



```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[5, 0],
              [10, 3]]) # A is the row player (in this case, prisoner)

B = np.array([[5, 10],
              [0, 3]]) # B is the column player
              # Form the game
game2 = nash.Game(A,B)
game2

     
```

```python
# Find the Nash Equilibrium with Support Enumeration
equilibria = game2.support_enumeration()
for eq in equilibria:
    print(eq)
     
```


**1.3 Elaborate on the Nash equilibrium solution of Model 1:**

The Nash Equilibrium in this competitive game is（no obvious ad，no obvious ad）.In this equilibrium both players choose not to release post with obvious advertisement which leads to a payoff 5 for each other.

A strategy is a Nash Equilibrium strategy if it is the best strategy for an agent, given the strategies the other players are actually using.

In this competitive game, neither player can unilaterally change their strategy to improve their own payoff, given the strategy of the other player. As a result, no matter what the other side chooses, not releasing post with obvious advertisement is always the best choice under the circumstances.

Therefore, in this case, the NE strategy is （no obvious ad，no obvious ad），both players end up with a suboptimal outcome.
